##Auto Loader

```python
(spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "parquet")
    .option("cloudFiles.schemaLocation","/Volumes/demoworkspace/default/bookstore_checkpoints/orders")
    .load("/Volumes/demoworkspace/default/bookstore_dataset/orders-raw")
.writeStream
    .option("checkpointLocation", "/Volumes/demoworkspace/default/bookstore_checkpoints/orders")
    .toTable("order_updates")
)
```

```python
%sql
select * from order_updates
```
