### Let explore what to do when data is accidentally deleted

```sql
delete from employees where id in(4,1) -- suppose we deleted few data
```

```sql
select * from employees
```

```sql
RESTORE TABLE practice.employees VERSION AS OF 5 -- Rollback to Version 5 before deletion of records
```

```sql
select * from employees
```

```sql
DESCRIBE HISTORY employees
```
