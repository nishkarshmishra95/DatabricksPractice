##CRAS Statement to override table

```sql
create or replace table orders as
select * from parquet.`/Volumes/demoworkspace/default/bookstore_dataset/orders`
```

```sql
describe history orders
```

## INSERT OVERWRITE statement to Overwrite data

```sql
insert overwrite orders
select * from parquet.`/Volumes/demoworkspace/default/bookstore_dataset/orders`
```

```sql
describe history orders
```

###Merge into command to prevent inserting same records multiple times

```sql
create or replace temporary view customers_updates as
select * from json.`/Volumes/demoworkspace/default/bookstore_dataset/customers-json-new`;

merge into customers c
using customers_updates u
on c.customer_id = u.customer_id
when matched and c.email is null and u.email is not null then
    update set email= u.email, updated=u.updated
when not matched then
    insert *
```

```sql
create or replace temp view books_updates 
 (book_id string,title string,author string,category string, price double)
using csv
options (
    path "/Volumes/demoworkspace/default/bookstore_dataset/books-csv-new", 
    header "true", 
    delimiter ";"
    );

select * from books_updates
```

```sql
merge into books b
using books_updates u
on b.book_id = u.book_id and b.title = u.title
when not matched and u.category = 'Computer Science' then
    insert *
```
