##Filter function

```sql
select 
    order_id,
    books,
    filter(books, i->i.quantity>=2) as multiple_copies
from orders
```

```sql
select order_id,multiple_copies from (
select 
    order_id,
    filter(books, i->i.quantity>=2) as multiple_copies
from orders)
where size(multiple_copies)>0;
```

##transform Function

```sql
select order_id,
books,
transform(
    books,
    b-> cast(b.subtotal*0.8 as int)
) as subtotal_discounted
from orders
```

## User Defined functions(UDF)

```sql
create or replace function get_url(email string)
returns string
return concat("https://www.",split(email,'@')[1])
```

```sql
select email,get_url(email) as domain from customers
```

```sql
describe function get_url
```

```sql
describe function extended get_url
```

```sql
create or replace function site_type(email string)
returns string
return case 
    when email like '%.com' then 'commercial business'
    when email like '%.org' then 'Non-profits Organization'
    when email like '%.edu' then 'Educational Institution'
    else concat("Unknown extension for domain: ",split(email,'@')[1])
    END
```

```sql
select email,site_type(email) as domain_category from customers
```

```sql
drop function if exists site_type;
drop function if exists get_url;
```
